```python
df=np.loadtxt('lab7_ex1_q.txt',dtype ='str')
```

```python
data=np.loadtxt('lab7_ex1_c.txt',dtype ='str')
```

```python
print(df)
```

```python
print(data)
```

```python
model=ConcreteModel()
```

```python
M = df.shape[1]-5
N=data.shape[0]-5
```

```python
row_indices=range(M)
col_indices=range(N)
print(row_indices)
print(col_indices)
```

```python
model.y=Var(range(8),range(8),range(8),range(8),domain=Binary)
model.x=Var(row_indices,col_indices,domain=Binary)
```

```python
model.constraint=ConstraintList()
```

```python
model.objective = Objective(expr=sum(int(data[j+1][l+1])*int(df[i+1][k+1])*(model.y[i,j,k,l])for i in row_indices for j in col_indices  for k in row_indices for l in col_indices),sense=minimize)
```

```python
for i in range(0,8):
   for j in range(0,8):
      for k in range(0,8):
        for l in range(0,8):
          model.constraint.add(expr=model.y[i,j,k,l]==model.y[k,l,i,j])
          
```

```python
for j in range(8):
  for k in range(8):
    for l in range(8):
      model.constraint.add(expr=sum(model.y[i,j,k,l] for i in range(8))==model.x[k,l])
for i in range(8):
  for k in range(8):
    for l in range(8):
      model.constraint.add(expr=sum(model.y[i,j,k,l] for j in range(8))==model.x[k,l])
```

```python
for i in row_indices:
  model.constraint.add(expr = sum(model.x[i,j] for j in col_indices)==1)
```

```python
for j in col_indices:
  model.constraint.add(expr=sum(model.x[i,j] for i in row_indices) == 1)
```

```python
model.pprint()
```

```python
!apt-get install -y -qq coinor-cbc
```

```python
opt_cbc= SolverFactory('cbc',)
```

```python
result = opt_cbc.solve(model)
print('solver statuas:',result.solver.status)
print('solver termination condition:',result.solver.termination_condition)
print('solver time taken:',result.solver.time)
```

```python
print('\nObjective=',model.objective())
for i  in row_indices:
  for j in col_indices:
    if model.x[i,j].value==1:
      print('Factory',i+1,'locate',j+1,':')
```
